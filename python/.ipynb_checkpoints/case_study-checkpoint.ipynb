```python
import pandas as pd 
```

```python
data = pd.read_csv('result_with_hp.csv')
```

```python
data.head()
```

```python
data['online_aTF'] = data['destination'] == data['online']
data['prediction_aTF'] = data['destination'] == data['prediction']
data['gx_aTF'] = data['destination'] == data['gx'].str.strip(' ')
data['lsl_aTF'] = data['destination'] == data['lsl'].str.strip(' ')
data['zyb_aTF'] = data['destination'] == data['zyb'].str.strip(' ')
```

```python
data['gx_aTF'].sum(),data['lsl_aTF'].sum(),data['zyb_aTF'].sum(),data['prediction_aTF'].sum() , data['online_aTF'].sum() 
```

```python
import re
import numpy as np
def data_clean(s):
    if s is np.nan : 
        return None 
    s1 = re.sub(r'[-. ·（ ）小区( ) 楼 酒店 宾馆 地铁 部 广场 小区 ]', '' , s)
    s2 = s1.rstrip('东门 ').rstrip('北门').rstrip('南门').rstrip('西门').rstrip('东北门')\
.rstrip('西北门').rstrip('西南门')
    return s2

from fuzzywuzzy import fuzz
def string_compare(s1,s2):
    return True if fuzz.partial_ratio(s1,s2) > 80 else False
```

```python
online_result = []
prediction_result = []
zyb_result = []
lsl_result = []
gx_result = []
label = data['destination'].apply(data_clean).values
online = data['online'].apply(data_clean).values
prediction = data['prediction'].apply(data_clean).values
gx = data['gx'].apply(data_clean).values
lsl = data['lsl'].apply(data_clean).values
zyb = data['zyb'].apply(data_clean).values

for i in range(len(label)):
    online_result.append(string_compare(label[i], online[i]))
    prediction_result.append(string_compare(label[i],prediction[i] ) ) 
    gx_result.append(string_compare(label[i],gx[i] ) ) 
    lsl_result.append(string_compare(label[i],lsl[i] ) ) 
    zyb_result.append(string_compare(label[i],zyb[i] ) ) 
```

```python
sum(zyb_result),sum(lsl_result),sum(gx_result) , sum(online_result)/150 , sum(prediction_result)/150
```

```python
data['online_TF'] = online_result
data['prediction_TF'] = prediction_result
data['gx_TF'] = gx_result
data['lsl_TF'] =  lsl_result
data['zyb_TF'] = zyb_result
```
