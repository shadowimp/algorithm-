````python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 

%matplotlib inline 


x = np.array([1,2,3])
y = np.array([2,4,6] )
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show() 
````

````python

s = pd.Series(np.random.rand(10).cumsum(),index=np.arange(0,100,10)) 

plt.plot(s,'r--')


plt.show() 
````

````python
x = np.array([1,2,3])
y = np.array([2,4,6] )
y2 = np.array([3,6,9])

plt.plot(x,y,color = 'g' , linestyle = '-' ) # 绿色 ， 虚线
plt.plot(x,y2,color = 'r', linestyle = '--') # 红色，虚线
plt.title('different color')

plt.show() 
````

### scatter 散点图

````python
x = np.array([1,2,3])
y = np.array([2,4,6] )
plt.scatter(x,y) #散点图
plt.show()
````

````python
x = np.random.rand(10)
y = np.random.rand(10)
plt.scatter(x,y,color='red')
plt.show()
````

### 柱状图

````python
data = [5, 20, 15, 25, 10]
labels = ['Tom', 'Dick', 'Harry', 'Slim', 'Jim']
plt.bar(range(len(data)), data, color ='grby' , tick_label = labels)   # 

plt.show()
````

```python
bar(left, height, width=0.8, bottom=None, **kwargs)
```
这四个参数确定了柱体的位置和大小。默认情况下，left为柱体的居中位置
- (left - width / 2, bottom)为左下角位置
- (left + width / 2, bottom + height)为右上角位置


### 条形图

````python
data = [5, 20, 15, 25, 10]

plt.barh(range(len(data)), data)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([5, 20, 15, 25, 10])
b = np.array([10, 15, 20, 15, 5])

plt.barh(range(len(a)), a)
plt.barh(range(len(b)), -b)
plt.show()
````

### 多张图

````python
x = [3,4,5]
y = [3,4,5]
plt.subplot(2,1,1) # 子图， 两行一列中的第一个 
plt.plot(x,y)

x1 = [1,2,3]
y1 = [5,6,7]
plt.subplot(2,1,2)  # 子图， 两行一列中的第二个 

plt.plot(x1,y1)
plt.show() 
````

## 正弦函数
$$ y = \sin(2\pi x ) $$

````python
x = np.linspace(0,1,100)
y = np.sin(2*np.pi*x)

plt.figure()
plt.plot(x,y)
plt.show() 
````

### seaborn 

````python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')

x = np.random.normal(1, 2, 5000)
y = np.random.normal(-1, 3, 2000)
bins = np.linspace(-10, 10, 30)

plt.hist([x, y], bins, label=['x', 'y'])
plt.legend(loc='upper right')
plt.show()
````

### pandas 直接plot画图

````python
iris_data = pd.read_csv('Iris.txt',header = None  ,index_col = None ,encoding = 'utf-8')
iris_data.columns = ['sepal_length','sepal_width','petal_length','petal_width','class']
iris_data.head()
````

````python
iris_data.plot(xticks = np.arange(0,150,10),figsize=(20, 5))  # figsize更改图像长宽
plt.show()
````

````python
plt.plot(iris_data['petal_length'])
plt.show()
````

````python
df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'],index=np.arange(10))
df.plot(xticks = np.arange(15),ylim=(0,2),alpha = 0.5 ) #xticks 确定x轴刻度  , ylim确定y轴界限  , alpha设置透明度，0为透明
# kind 确定图像类型，  line折线图， bar柱状图，barh横向柱状图， kde密度图
plt.grid()  # grid添加网格

plt.show()

````

## hist 直接画直方图

````python
data = pd.Series(np.array([1,3,4,5,4]),index = np.arange(5))
data.hist(alpha= 0.5)  
plt.title('hist')
plt.xlabel('type')
plt.ylabel('number')
plt.legend('person')
plt.show()
````
