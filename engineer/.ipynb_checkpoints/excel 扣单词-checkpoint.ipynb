```python
import xlrd

book = xlrd.open_workbook("aaa.xls")
sh = book.sheet_by_index(0)
print("{0} {1} {2}".format(sh.name, sh.nrows, sh.ncols))


import re
def twoSum(sentence,word_order):
    # sentence = "i love you"    
    f = lambda x: [e.lower() for e in re.findall(r'\b\w+\b', x)]
    a = f(sentence)
    # word_order = 2
    word = a[word_order-1]
    # print word
    return word 
twoSum("i ldidasd dssdff sdf",3)

# print '*'*10
# mistake = sh.cell_value(rowx=1,colx=3)
# print mistake

words = []
mistake = []
correct = []
for i in range(1,sh.nrows):
    sentence= sh.cell_value(rowx=i, colx=1)
    sentence_fix = sentence.replace('<br/>','')
    sentence_fix2 = sentence_fix.replace("'",'')
    word_order = int(sh.cell_value(rowx=i, colx=2) )
    words.append(twoSum(sentence_fix2,word_order))
    mistake.append (sh.cell_value(rowx=i,colx=3))
    correct.append(1 if words[i-1]==mistake[i-1] else 0)
print correct

print words


import xlwt
from datetime import datetime
style0 = xlwt.easyxf('font: name Times New Roman, color-index red, bold on',
    num_format_str='#,##0.00')
style1 = xlwt.easyxf(num_format_str='D-MMM-YY')
wb = xlwt.Workbook()
ws = wb.add_sheet('A Test Sheet')

# ws.write(21, 4, 1234.56)

# ws.write(2, 0, 1)
# ws.write(2, 1, 1)
# ws.write(3, 2, 4)
# ws.write(2, 2, xlwt.Formula("A3+B3"))

for i in range(0,600):
	ws.write(i, 1, words[i])
	ws.write(i, 2, mistake[i])
	ws.write(i, 3, correct[i])


wb.save('example3.xls')
```
